```python
train_df = pd.DataFrame({'Degree': ['Masters', 'PHD', 'High School', 'Diploma', 'Masters', 'Bachelors', 'High School', 'PHD', 'Bachelors', 'Diploma']})
#train_df

# create object of ordinal encoding
encoder = ce.OrdinalEncoder(cols=['Degree'], return_df=True, mapping = [
        {'col': 'Degree', 'mapping': {None: 0, 'High School': 1, 'Diploma': 2,  'Bachelors': 3,  'Masters': 4,  'PHD': 5}},
        ])

#Original data
train_df
```

# one hot coding

```python
import pandas as pd
import category_encoders as ce
 
```

```python
data = pd.DataFrame({'Colors': ['red', 'yellow', 'green', 'blue','red', 'yellow', 'green', 'blue']})


#create object for one hot encoding
encoder = ce.OneHotEncoder(cols= 'Colors', handle_unknown= 'return_nan', return_df=True, use_cat_names=True)


# original data
data
```

```python
#fit transform train data
data_transformed = encoder.fit_transform(data)
data_transformed
```

# Dummy encoding

```python
data_transformed = pd.get_dummies(data = data, drop_first=True)
data_transformed
```

# Effect encoding

```python
encoder = ce.sum_coding.SumEncoder(cols='Colors')
encoder.fit_transform(data)
```

# Hashing Encoding

```python
#create object for hash encoder
encoder = ce.HashingEncoder(cols='Colors', n_components=6)
```

```python
#fit and transform the data
encoder.fit_transform(data)
```

# Binary Encoding

```python
#create object for binary encoding
encoder = ce.BinaryEncoder(cols='Colors', return_df=True)
```

```python
#fit and transform the data
encoder.fit_transform(data)
```
